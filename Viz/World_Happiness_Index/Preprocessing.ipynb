## Standardizing the Columns

### The Dysotopia Residual Column

```python
dfs = [fifteen, sixteen, seventeen, eighteen, nineteen, twenty, twentyone, twentytwo]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
assert len(dfs) == len(years)
```

```python
for year, df in zip(years, dfs):
    print(year, df.shape)
```

So, for year 2018 and 2019, we don't have the column 'Dystopia Residual'. Apart from that, the columns are the same in each dataset. The order is different in each dataset. We'll need to standardize the columns.

The order of columns will be:
1. Country
2. Region
3. Happyness Rank
4. Happyness Score
5. Economy
6. Family
7. Health
8. Freedom
9. Generosity
10. Corruption
11. Dystopia Residual 

```python
standard_columns = [
    "Country",
    "Region",
    "Happiness Rank",
    "Happiness Score",
    "Economy",
    "Family",
    "Health",
    "Freedom",
    "Generosity",
    "Trust",
    "Dystopia",
]
```

To work with all the datasets, we'll add a columns `Dystopia` to the 2018 and 2019 datasets. We'll fill the values with 0.

```python
eighteen["Dystopia"] = 0
nineteen["Dystopia"] = 0
```

```python
for year, df in zip(years, dfs):
    print(year, df.shape)
```

### Standardizing the Column Names

```python
df_finals = []
for df in dfs:
    df = df[standard_columns]
    df_finals.append(df)
```

```python
for df in df_finals:
    if df.columns.tolist() != standard_columns:
        print("Not equal")
```

Great! Work done! We just need to drop the columns which we added earlier.

```python
df_finals[years.index(2018)].drop("Dystopia", axis=1, inplace=True)
df_finals[years.index(2019)].drop("Dystopia", axis=1, inplace=True)
```

## Creating a Single Dataset

```python
for df in df_finals:
    print(df.shape)
```

Let's save the datasets.

```python
for df, year in zip(df_finals, years):
    df.to_csv(f"final/{year}.csv", index=False)
```
