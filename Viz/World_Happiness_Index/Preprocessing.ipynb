## Standardizing the Data

### Fifteen

```python
fifteen.columns
```

Let's drop the extra column.

```python
fifteen.drop(columns=["Standard Error"], axis=1, inplace=True)
```

Now, the columns of this dataset will be the standard columns.

```python
standard_columns = fifteen.columns.tolist()
```

### Sixteen

```python
sixteen.columns
```

We'll again drop the extra columns

```python
sixteen.drop(columns=['Lower Confidence Interval', 'Upper Confidence Interval'], axis=1, inplace=True)
```

```python
def compare_columns(columns):
    for column in columns:
        if column not in standard_columns:
            return False
    return True
```

```python
compare_columns(sixteen.columns)
```

Next, we'll merge all these datasets into one.

```python
fifteen= pd.read_csv("2015_final.csv")
sixteen= pd.read_csv("2016_final.csv")
seventeen= pd.read_csv("2017_final.csv")
eighteen= pd.read_csv("2018_final.csv")
nineteen= pd.read_csv("2019_final.csv")
twenty= pd.read_csv("2020_final.csv")
twentyone= pd.read_csv("2021_final.csv")
twentytwo= pd.read_csv("2022_final.csv")
```

```python
df_temp = pd.merge(fifteen, sixteen, on='Country', suffixes=['_2015', '_2016'])
df_temp_2 = pd.merge(seventeen, eighteen, on='Country', suffixes=['_2017', '_2018'])
df_temp_3 = pd.merge(nineteen, twenty, on='Country', suffixes=['_2019', '_2020'])
df_temp_4 = pd.merge(twentyone, twentytwo, on='Country', suffixes=['_2021', '_2022'])
```

```python
df_temp_5 = pd.merge(df_temp, df_temp_2, on='Country')
df_temp_6 = pd.merge(df_temp_5, df_temp_3, on='Country')
df_temp_7 = pd.merge(df_temp_6, df_temp_4, on='Country')
df_temp_7.columns
```
