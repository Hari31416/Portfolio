# Viz

## Happiness Score of Countries

We'll be plotting a world map showing the happiness score of countries from year 2015 to 2022. This will give us an idea of the global happiness trend.

```python
def plot_happiness_score(year):
    year = str(year)
    fig = px.choropleth(df, locations="Code",
                    color=f"Happiness Score_{year}",
                    hover_name="Country",
                    color_continuous_scale=px.colors.sequential.Plasma,
                    hover_data=[f"Happiness Rank_{year}"],
                    title=f"World Happiness Score in {year}")
        
    fig.show()
```

```python
interact(plot_happiness_score, year=(2015, 2022),);
```

You can see the trend of happiness score of countries in the map above by using the slider on top of the map.

```python
plot_happiness_score(2015)
```
