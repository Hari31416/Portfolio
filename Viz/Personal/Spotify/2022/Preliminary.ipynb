```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("MyData/StreamingHistoryExtras.csv")
```

```python
df.head()
```

```python
df["endTime"] = pd.to_datetime(df["endTime"])
```

```python
minutes_played = df["Minutes Played"].sum()
hours_played = minutes_played / 60
distinct_artist = df["artistName"].nunique()
distinct_songs = df["trackName"].nunique()
total_songs = df["trackName"].count()

print("Total minutes played: ", minutes_played)
print("Total hours played: ", hours_played)
print("Total distinct artists: ", distinct_artist)
print("Total distinct songs: ", distinct_songs)
print("Total songs: ", total_songs)
```

```python
start_time = df["endTime"].min()
end_time = df["endTime"].max()
total_days = (end_time - start_time).days
total_hours = (end_time - start_time).total_seconds() / 3600

percentage_time = (hours_played / total_hours) * 100
print(f"Percentage of time spent listening to music: {percentage_time:.2f}%")
```
