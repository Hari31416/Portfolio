```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("goodreads_library_export.csv")
```

```python
df.columns
```

```python
df = df[['Book Id', 'Title', 'Author', 'ISBN', 'ISBN13', 'My Rating', 'Average Rating', 'Publisher',
    'Number of Pages', 'Year Published', 'Original Publication Year',
       'Date Read', 'Date Added', 'Bookshelves','Read Count']]
```

```python
df["Date Read"] = pd.to_datetime(df["Date Read"])
df["Date Added"] = pd.to_datetime(df["Date Added"])
```

```python
df.head()
```

```python
df.info()
```

```python
df['ISBN'] = df["ISBN"].apply(lambda x : x[2:-1])
df['ISBN13'] = df["ISBN13"].apply(lambda x : x[2:-1])
```

```python
df["ISBN"] = df["ISBN"].apply(lambda x : x if x != df["ISBN"][0] else np.nan)
df["ISBN13"] = df["ISBN13"].apply(lambda x : x if x != df["ISBN13"][0] else np.nan)
```

```python
df.head()
```

```python
df.info()
```

```python
df["Bookshelves"].fillna("read", inplace=True)
```

```python
def update_shelve(row):
    count = row["Read Count"]
    count = int(count)
    shelves = row["Bookshelves"]
    if shelves == "read":
        return shelves
    if count:
        shelves += ", read"
    return shelves
```

```python
df["Bookshelves"] = df[["Bookshelves", "Read Count"]].apply(update_shelve, axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv("All/books_cleaned.csv", index=False)
```

## Data Preprocessing

### Categorizing into Series

```python
import re
def get_series_name(name):
    pattern = r"\((.*),.*?\)"
    match = re.search(pattern, name)
    if match:
        return match.group(1)
    else:
        return "Stand Alone"
```

```python
df["Series"] = df["Title"].apply(lambda x : get_series_name(x))
```
