```python
import pandas as pd
import numpy as np
import numpy as np
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/The_Wheel_of_Time")
```

```python
books_raw = df[1]
```

```python
books_raw["Date"] = pd.to_datetime(books_raw["Date"])
```

```python
books_raw.drop("Notes", axis=1, inplace=True)
```

```python
books_raw.drop(15, axis=0, inplace=True)
```

```python
books_raw["Pages"] = books_raw["Length"].apply(lambda x: int(x.split("pp")[0]))
```

```python
import re
def num_words(x):
    regex = re.compile(r"\d{3},\d{3}")
    return int(regex.findall(x)[0].replace(",", ""))
```

```python
books_raw["Words"] = books_raw["Length"].apply(num_words)
```

```python
books_raw.drop("Length", axis=1, inplace=True)
```

```python
books_raw.to_csv("WoT.csv", index=False)
```
