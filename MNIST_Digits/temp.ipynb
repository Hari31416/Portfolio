```python
def has_33(nums):
    text = "".join(str(x) for x in nums)
    return "33" in text
```

```python
has_33([1, 3, 3]), has_33([1, 3,1, 3])
```

```python
def has_33(nums):
    for i in range(len(nums)-1):
        if nums[i] == 3 and nums[i+1] == 3:
            return True
    return False
```

```python
has_33([1, 3, 3]), has_33([1, 3,1, 3])
```

```python
for i in range(5):
    print(i)
    if i>3:
        print("i is greater than 3. Breaking")
        break
else:
    print("This is the else block")
```

```python
i=0
while i<5:
    print(i)
    if i>3:
        print("i is now greater than 3. Breaking out of loop")
        break
    i+=1
else:
    print("i is no longer less than 5")
```
