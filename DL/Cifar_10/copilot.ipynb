```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

hcut=0.4         
a=1
def HAB(kx, ky):
    H_ab= hcut*((np.exp(1j*(((kx)*a/2)+((ky)*np.sqrt(3)/2))))  
                 + (np.exp(1j*(((kx)*a/2)-((ky)*np.sqrt(3)/2)))) 
                + (np.exp(-1j*((kx)*a)))) 
    return H_ab 

kx= np.linspace(-2*np.pi/a, 2*np.pi/a, 100, dtype=float)
ky= np.linspace(-2*np.pi/a, 2*np.pi/a, 100, dtype=float)
E1=[0]*len(kx)
E2=[0]*len(kx)

for i in range(0,len(kx)):
    H_ab= HAB(kx[i], ky[i])                               # fixing potential value
    H_ba= H_ab.conjugate()                    
    
   #creating matrix of hamiltonian
    A= np.array([[0 , H_ab],
                 [H_ba , 0]])
    
    # finding eigenvalue and eigenvectors
    Eigen_E, Eigen_V= np.linalg.eig(A)
    
    E1[i]=Eigen_E[0]
    E2[i]=Eigen_E[1]

E1= np.array(E1).real
E2= np.array(E2).real


x, y = np.meshgrid(kx, ky)
E12 = E1.reshape(-1, len(kx))
E22 = E2.reshape(-1, len(kx))
fig = plt.figure(figsize =(14, 9))
ax = fig.add_subplot(projection='3d')

# Creating plot
ax.plot_surface(x, y, E12, cmap='viridis')
ax.plot_surface(x, y, E22, cmap= "gnuplot")

ax.view_init(elev=10, azim=80)
plt.title('E vs K diagram (Reduced Zone Scheme)')
```

```python
hbar = 1
def return_HAB(kx,ky,a):
    t1 = exp(1j*((kx*a/2)+ky*(sqrt(3)*a/2)))
    t2 = exp(1j*(((kx*a/2)-(ky*sqrt(3)/2))))
    t3 = exp(-1j*(kx*a))
    H_ab= hcut*(t1+t2+t3)
    H_ab.subs(kx,kx)
    H_ab.subs(ky,ky)
    H_ab.subs(a,a)
    H_ab = H_ab.evalf()
    return H_ab*hbar
```

```python
def HAB(kx, ky,a=1, hcut = 0.4):
    t1 = np.exp(1j*(kx*a/2)+ky*(np.sqrt(3)*a/2))
    t2 = (np.exp(1j*(((kx)*a/2)-((ky)*np.sqrt(3)/2))))
    t3 = (np.exp(-1j*((kx)*a)))
    return hcut*(t1+t2+t3)
```

```python
kx= np.linspace(-2*np.pi/a, 2*np.pi/a, 100, dtype=float)
ky= np.linspace(-2*np.pi/a, 2*np.pi/a, 100, dtype=float)
E1=[0]*len(kx)
E2=[0]*len(kx)

for i in range(len(kx)):
    H_ab= HAB(kx[i], ky[i])
    mat = np.array([[0, H_ab], [H_ab.conjugate(), 0]])
    Eigen_E, Eigen_V= np.linalg.eig(mat)
    E1[i]=Eigen_E[0]
    E2[i]=Eigen_E[1]
```
