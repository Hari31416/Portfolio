```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
labels = pd.read_csv("data/trainLabels.csv")
labels.head()
```

```python
labels.groupby("label").count()
```

```python
labels_list = labels["label"].unique().tolist()
```

```python
for label in labels_list:
    directory = os.path.join("data/train", label)
    if not os.path.exists(directory):
        os.makedirs(directory)
        
```

```python
images = os.listdir("data/train")
images = [image for image in images if image.endswith(".png")]
len(images)
```

```python
directories = os.listdir("data/train")
directories = [directory for directory in directories if not directory.endswith(".png")]
len(directories)
```

```python
def get_image_category(image):
    image_id = int(image.split(".")[0])
    return labels[labels["id"] == image_id]["label"].values[0]
```

```python
for image in images:
    category = get_image_category(image)
    directory = os.path.join("data/train", category)
    shutil.move(os.path.join("data/train", image), os.path.join(directory, image))
```

```python
for dir in directories:
    print(len(os.listdir(os.path.join("data/train", dir))))
```

```python
for dir in directories:
    os.mkdir(os.path.join("data/dev", dir))
```

```python
import tqdm
from IPython.display import clear_output
```

```python
DEV_SPLIT = 0.04

np.random.seed(42)

for dir in tqdm.tqdm(directories):
    print(f"Copying images from {dir}")
    image_dir = os.path.join("data/train", dir)
    images = os.listdir(image_dir)
    num_images = len(images)
    num_dev_images = int(DEV_SPLIT * num_images)
    np.random.shuffle(images)
    dev_images = images[:num_dev_images]
    for image in dev_images:
        dest_dir = os.path.join("data/dev", dir)
        shutil.move(os.path.join(image_dir, image), os.path.join(dest_dir, image))
    print(f"Done the {dir}!")
```
