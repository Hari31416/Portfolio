```python
def has_33(nums):
    text = "".join(str(x) for x in nums)
    return "33" in text
```

```python
def has_33(nums):
    for i in range(len(nums)-1):
        if nums[i] == 3 and nums[i+1] == 3:
            return True
    return False
```

```python
def user_input(within_range = range(3)):
    choice = False
    acceptable_range = False
    while not choice or not acceptable_range:
        choice = input(f"Please input a number within {within_range[0]} and {within_range[-1]} ")

        if not choice.isdigit():
            print("Please enter a number.")
        else:
            if int(choice) in within_range:
                acceptable_range = True
            else:
                print("Number not in the range")
    return int(choice)
```
