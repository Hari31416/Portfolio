## Definiation

Fresnel Integrals are defined as:
$$C(\tau) = \int_0^\tau \cos{(\frac{1}{2}\pi u^2)} du$$
$$S(\tau) = \int_0^\tau \sin{(\frac{1}{2}\pi u^2)} du$$

```python
from sympy import *
from sympy.abc import x, y, z
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fresnel_c(c):
    return cos(1/2*pi*c**2)

def fresnel_s(c):
    return sin(1/2*pi*c**2)
```

```python
def integrate(f, b, a=0):
    return Integral(f(x), (x, a, b)).doit().evalf()

integrate = np.vectorize(integrate)
```

```python
nums = np.arange(0.1, pi, 0.01)
fresnel_c_i = integrate(fresnel_c, nums)
fresnel_s_i = integrate(fresnel_s, nums)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(nums, fresnel_c_i, label="C(x)")
plt.plot(nums, fresnel_s_i, label='S(x)')
plt.legend();
```
