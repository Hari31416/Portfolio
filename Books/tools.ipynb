```python
import ebooklib
import bs4
from ebooklib import epub
from tqdm import tqdm
import mobi
import os
import shutil
```

```python
def word_count_epub(path):
    """
    Counts the number of words in an epub file.

    Attributes:
    -------
    path : str
        Path to the epub file.
    
    Returns
    -------
    int
        Number of words in the epub file.
    """
    book = epub.read_epub(path)
    i = 0
    count = 0
    chapters = list(book.get_items_of_type(ebooklib.ITEM_DOCUMENT))
    for item in tqdm(chapters, desc="Calculating..."):
        if item.is_chapter():
            html = item.get_body_content().decode('utf-8')
            soup = bs4.BeautifulSoup(html, 'html.parser')
            text = soup.text
            count += len(text.split())
            i += 1
    return count
```

```python
def word_count_mobi(file):
    """
    Counts the number of words in a mobi file.

    Attributes:
    -------
    path : str
        Path to the mobi file.
    
    Returns
    -------
    int
        Number of words in the mobi file.
    """
    file_dir, filepath = mobi.extract(file)
    soup = bs4.BeautifulSoup(open(filepath, encoding="utf-8"), 'html.parser')
    words = len(soup.text.split())
    os.remove(filepath)
    shutil.rmtree(file_dir)
    return words
```
