## Neural Networks

```python
type(X_train)
```

```python
X_input = tfl.Input((286), name="input")
X = tfl.Flatten()(X_input)
X = tfl.Dense(32, activation="relu", name="Dense1")(X)
X = tfl.Dense(64, activation="relu", name="Dense2")(X)
X = tfl.Dense(64, activation="relu", name="Dense3")(X)
X = tfl.Dense(128, activation="relu", name="Dense4")(X)
X = tfl.Dense(128, activation="relu", name="Dense5")(X)
X = tfl.Dense(256, activation="relu", name="Dense6")(X)

output = tfl.Dense(10, activation="softmax", name="output")(X)
model = Model(X_input, output, name="model_1")
model.summary()
```

```python
model.compile(optimizer='adam', loss = "categorical_crossentropy", metrics=["accuracy"])
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```
