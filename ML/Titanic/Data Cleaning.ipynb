## Taking Care of Missing Values

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

`Age` and `Cabin` have quite a few missing values for both training and test set. Also, in test set, one value of `Fare` is also missing. Let's start with `Age`.

### Age Column

```python
train['Age'].describe()
```

One way to handle missing values is to replace them with the mean of the column. However, the ages have a large standard deviation. So, replacing the missing values with the mean of the column is not a good idea. Let's see if we can find a better way to handle missing `Age` values.

```python
train[train['Age'].isnull()]
```

```python
train[train['Age']<5]
```

It seems that children has 'Master' or 'Miss' as their name. This might be useful while filling the missing age values.

Let's see the ages of passangers with 'Master' or 'Miss' names.

```python
train[train['Name'].apply(lambda x: 'master' in x.lower())].sort_values('Age', ascending=False).head(10)
```

**The maximum value of age for a passenger with name containig 'Master' is 12.**  This means that if that we can easly fill the missing age for these types of passengers with the mean. Let's calculate it and use it to fill the missing age values.

```python
#Ues mean value of all the passengers' age with name containing master to fill corresponding null values
mean_master_names = np.round(train[train['Name'].apply(lambda x: 'master' in x.lower())]['Age'].mean(),2)
master_named_data = train[train['Name'].apply(lambda x: 'master' in x.lower())]
master_named_data["Age"].fillna(mean_master_names, inplace=True)
train.iloc[master_named_data.index] = master_named_data
```

```python
train.isnull().sum()
```

Let's do the same to the test data.

```python
test.isnull().sum()
```

```python
mean_master_names = np.round(test[test['Name'].apply(lambda x: 'master' in x.lower())]['Age'].mean(),2)
master_named_data = test[test['Name'].apply(lambda x: 'master' in x.lower())]
master_named_data["Age"].fillna(mean_master_names, inplace=True)
test.iloc[master_named_data.index] = master_named_data
```

```python
test.isnull().sum()
```

```python
miss_named_data = train[train['Name'].apply(lambda x: 'miss' in x.lower())].sort_values('Age', ascending=False)
miss_named_data
```

```python
miss_named_data.describe()['Age']
```

The above method does not work that effectively in case the name has 'Miss' since the maximum value of age for a passenger with name containig 'Miss' is 63! And standard deviation is also very high.

Let's see if there are some other relationships we can use. The columns `SibSp` and `Parch` seems to be interseting.

```python
mean_values = miss_named_data.groupby(['SibSp', 'Parch']).mean()['Age'].reset_index()
mean_values["Age"] = mean_values["Age"].round(2)
mean_values
```

Okay, this is much better. Let's use this!

```python
def fill_age_na(row):
    n_sibsp = row['SibSp']
    n_parch = row['Parch']
    age = row['Age']
    for i,j, k in zip(mean_values['SibSp'].values, mean_values['Parch'].values, mean_values['Age'].values):
        if i==n_sibsp and j==n_parch and np.isnan(age):
            row['Age'] = k
            return row
    return row

df_temp_train = miss_named_data.apply(lambda x: fill_age_na(x), axis=1)
train.iloc[df_temp_train.index] = df_temp_train
```

```python
train.isnull().sum()
```

Let's do the same thing to the test set.

```python
miss_named_data = test[test['Name'].apply(lambda x: 'miss' in x.lower())].sort_values('Age', ascending=False)
mean_values = miss_named_data.groupby(['SibSp', 'Parch']).mean()['Age'].reset_index()
mean_values["Age"] = mean_values["Age"].round(2)
df_temp_test = miss_named_data.apply(lambda x: fill_age_na(x), axis=1)
test.iloc[df_temp_test.index] = df_temp_test
test.isnull().sum()
```

This concludes the subset of the dataframe where passanger name contained *Miss*. What about other values? Let's see if we can find some more pattern, if not, the same method as above can be applied to the whole train and test set.

```python
train[train['Age'].isna()]
```

Since there is no obvious pattern, we'll just groupby 'SibSp', 'Parch', 'Sex' and use their mean value to fill the null age values.

```python
mean_ages = train.groupby(['SibSp', 'Parch', 'Sex']).mean()['Age'].reset_index()
mean_ages["Age"] = mean_ages["Age"].round(2)
mean_ages
```

```python
def fill_age_na_2(row):
    n_sibsp = row['SibSp']
    n_parch = row['Parch']
    sex = row["Sex"]
    age = row['Age']
    for i,j, gender, k in zip(mean_ages['SibSp'].values, mean_ages['Parch'].values, mean_ages['Sex'].values, mean_ages['Age'].values):
        if i==n_sibsp and j==n_parch and gender == sex and np.isnan(age):
            row['Age'] = k
            return row
    return row
```

```python
train = train.apply(lambda x: fill_age_na_2(x), axis=1)
test = test.apply(lambda x: fill_age_na_2(x), axis=1)
```

Train set contains a null `Embarked` value. We'll use the most common value to fill it.

Test set also contains a missing `Fare` value. We can use the mean value to fill this, but let's see if we can find a better way.

```python
train['Embarked'].value_counts()
```

The `Fare` values ranges a lot. It won't be a good idea to fill the missing values with the mean value.

```python
test[test['Fare'].isnull()]
```

```python
test.groupby(['Pclass', 'Cabin']).mean()["Fare"]
```

```python
test.groupby(['Pclass', 'Embarked']).mean()["Fare"]
```

Using these two tables, we can take the mean of the mean fare value corresponding the classes. This gives.

```python
mean_fare = np.round((12.511463+13.913030)/2,2)
mean_fare
```

Let's use this to fill the missing `Fare` values in the test set.

```python
test['Fare'].fillna(mean_fare, inplace=True)
```
