```python
def get_mean1(nums):
    """ Gets the mean of the list of numbers
    
    Args:
        nums (list): the list of numbers
    
    Returns:
        float: the mean of the list of numbers
    """
    return sum(nums)/len(nums)
```

```python
def get_mean2(nums):
    """ Gets the mean of the list of numbers
    
    Args:
        nums (list): the list of numbers
    
    Returns:
        float: the mean of the list of numbers
    """
    return sum(nums)/max(len(nums),1)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("time_table.txt", delimiter="\t")
```

```python
df
```

```python
df["Date"] = df["Time"].apply(lambda x: x.split("/")[0])
df["Time"] = df["Time"].apply(lambda x: x.split("/")[1])
```

```python
def add(string):
    return str(string)+ " Hrs"
```

```python
df["Time"] = df["Time"].apply(add)
```

```python
df.sort_values(by=["Date", "Time"])
```

```python
df.groupby("Date").count()
```
