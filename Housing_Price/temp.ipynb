```python
def get_mean1(nums):
    """ Gets the mean of the list of numbers
    
    Args:
        nums (list): the list of numbers
    
    Returns:
        float: the mean of the list of numbers
    """
    return sum(nums)/len(nums)
```

```python
def get_mean2(nums):
    """ Gets the mean of the list of numbers
    
    Args:
        nums (list): the list of numbers
    
    Returns:
        float: the mean of the list of numbers
    """
    return sum(nums)/max(len(nums),1)
```
