```python
def remove_chars_from_folder(directory):
    for folder, _, _ in os.walk(directory):
        print(folder)
        try:
            folder_new = folder.replace(' ', '_').replace(".", "_")
            os.rename(folder, folder_new)
        except:
            print('Something went wrong with folder: ', folder)
```

```python
remove_chars_from_folder(directory)
```

```python
def remove_chars_from_file(directory, formats = ['ipynb']):
    for folder, _, files in os.walk(directory):
        for file in files:
            for format in formats:
                if file.endswith(format):
                    try:
                        file_new = file.replace(' ', '_')
                        old_file = os.path.join(folder, file)
                        new_file = os.path.join(folder, file_new)
                        os.rename(old_file, new_file)
                    except:
                        print('Something went wrong with file: ', file)
```

```python
remove_chars_from_file(directory)
```

### Converting to HTML

#### Adding Sharing URL

```python
def convert_notebook(notebook_path):
    notebook = nbformat.read(notebook_path, as_version=4)
    html_exporter = HTMLExporter(template_name="classic")
    colab_url = f"https://colab.research.google.com/github/Hari31416/Portfolio/blob/main/{notebook_path}?hl=en"
    github_url = f"https://github.com/Hari31416/Portfolio/blob/main/{notebook_path}"
    source = f"""<table class="tfo-notebook-buttons" align="left">
<td style="border: none;">
<a target="blank" href="{colab_url}"><img src="https://colab.research.google.com/img/colab_favicon_256px.png" width="28" height="28"/>Run on Google Colab</a></td>
<td style="border: none;">
<a target="blank" href="{github_url}"><img src="https://cdn.icon-icons.com/icons2/2368/PNG/512/github_logo_icon_143772.png" width="28" height="28"/>View on Github</a></td>
<td style="border: none;">
</table>
"""
    notebook['cells'][0]['source'] = source

    (body, _) = html_exporter.from_notebook_node(notebook)

    with open(notebook_path.replace('.ipynb', '.html'), 'w', encoding="utf-8") as f:
        #The Title
        title = notebook_path.split(os.path.sep)[-1].split(".")[0]
        title = title.replace("_", " ").title()
        body = re.sub("<title>Notebook</title>", f"<title>{title}</title>", body)
        #The Source
        f.write(body)

        return notebook_path.replace('.ipynb', '.html')
```

#### Converting

```python
def convert_to_html(directory):
    outputs = []

    for folder, _, files in os.walk(directory):
        if 'checkpoint' in folder:
            continue
        
        for file in files:
            if file.endswith('.ipynb'):
                notebook_path = os.path.join(folder, file)
                output = convert_notebook(notebook_path)
                outputs.append(output)
                print("Converted: ", file)
    return outputs
```

```python
outputs = convert_to_html(directory=directory)
```

## Adding Sharing Url

```python
def convert_notebook(notebook_path):
    notebook = nbformat.read(notebook_path, as_version=4)
    html_exporter = HTMLExporter(template_name="classic")
    colab_url = f"https://colab.research.google.com/github/Hari31416/Portfolio/blob/main/{notebook_path}?hl=en"
    github_url = f"https://github.com/Hari31416/Portfolio/blob/main/{notebook_path}"
    source = f"""<table class="tfo-notebook-buttons" align="left">
<td style="border: none;">
<a target="blank" href="{colab_url}"><img src="https://colab.research.google.com/img/colab_favicon_256px.png" width="28" height="28"/>Run on Google Colab</a></td>
<td style="border: none;">
<a target="blank" href="{github_url}"><img src="https://cdn.icon-icons.com/icons2/2368/PNG/512/github_logo_icon_143772.png" width="28" height="28"/>View on Github</a></td>
<td style="border: none;">
</table>
"""
    notebook['cells'].insert(0, nbformat.v4.new_markdown_cell(source))

    (body, _) = html_exporter.from_notebook_node(notebook)

    with open(notebook_path.replace('.ipynb', '.html'), 'w', encoding="utf-8") as f:
        #The Title
        title = notebook_path.split(os.path.sep)[-1].split(".")[0]
        title = title.replace("_", " ").title()
        body = re.sub("<title>Notebook</title>", f"<title>{title}</title>", body)
        #The Source
        f.write(body)

        return notebook_path.replace('.ipynb', '.html')
```
