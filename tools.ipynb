## Changing All notebooks to HTML

### Remove Spaces from Folder Names

```python
def remove_chars_from_folder(directory):
    for folder, _, _ in os.walk(directory):
        print(folder)
        try:
            folder_new = folder.replace(' ', '_').replace(".", "_")
            os.rename(folder, folder_new)
        except:
            print('Something went wrong with folder: ', folder)
```

```python
remove_chars_from_folder("ML\DataCamp")
```

```python
def remove_chars_from_file(directory, formats = ['ipynb']):
    for folder, _, files in os.walk(directory):
        for file in files:
            for format in formats:
                if file.endswith(format):
                    try:
                        file_new = file.replace(' ', '_')
                        old_file = os.path.join(folder, file)
                        new_file = os.path.join(folder, file_new)
                        os.rename(old_file, new_file)
                    except:
                        print('Something went wrong with file: ', file)
```

```python
remove_chars_from_file("ML\DataCamp")
```

### Converting to HTML

```python
def convert_to_html(directory):
    outputs = []
    n = Notebook(directory)
    for folder, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.ipynb'):
                output = n.export_notebook(os.path.join(folder, file))
                outputs.append(output)
                print("Converted: ", file)
    return outputs
```

```python
outputs = convert_to_html("ML\DataCamp")
```

### Rename all the files to index.html

#### Remove Previous index.html files

```python
directory  = "ML\DataCamp"
for folder, _, files in os.walk(directory):
    for file in files:
        if file == 'index.html':
            print("Removing from: ", folder)
            os.remove(os.path.join(folder, file))
```

```python
for file in outputs:
    file_name = file.split('\\')[-1]
    os.rename(file, file.replace(file_name, 'index.html'))
```
