## Making a text editor using tkinter

```python
import tkinter as tk
from tkinter.filedialog import askopenfilename, asksaveasfilename

#Functions
#Opening File
def open_file():
    """Open a file for editing."""
    global filepath
    filepath = askopenfilename(
        filetypes=[("Text Files", "*.txt"), ("All Files", "*.*")]
    )
    if not filepath:
        return
    text_editor.delete("1.0", tk.END)
    with open(filepath, "r") as input_file:
        text = input_file.read()
        text_editor.insert(tk.END, text)
    window.title(f"Text Editor - {filepath}*")
    return filepath
#Saving File
def save_file():
    """Save the current file as a new file."""
    global filepath
    filepath = asksaveasfilename(
        defaultextension="txt",
        filetypes=[("Text Files", "*.txt"), ("All Files", "*.*")],
    )
    if not filepath:
        return
    with open(filepath, "w") as output_file:
        text = text_editor.get("1.0", tk.END)
        output_file.write(text)
    window.title(f"Text Editor - {filepath}")
    return filepath

#Updating the file
def update_file():
    """Update the current file."""
    try:
        text = text_editor.get("1.0", tk.END)
        with open(filepath, "w") as output_file:
            output_file.write(text)
        window.title(f"Text Editor - {filepath}")
    except:
        tk.messagebox.showerror("Error", "No file is open.")

#The Window
window = tk.Tk()
window.title("Text Editor")
window.rowconfigure(0, minsize=800, weight=1)
window.columnconfigure(1, minsize=800, weight=1)

#Frame
frm_btn = tk.Frame(window)
frm_btn.grid(row=0, column=0, sticky="ns")

#The Editor
text_editor = tk.Text(window)
text_editor.grid(row=0, column=1, sticky="nsew")

#The Buttons
btn_open = tk.Button(frm_btn, text="Open", command = open_file)
btn_open.grid(row=0, column=0, sticky="ew", padx=5, pady=5)

btn_save = tk.Button(frm_btn, text="Save As...", command=save_file)
btn_save.grid(row=1, column=0, sticky="ew", padx=5, pady=5)

btn_update = tk.Button(frm_btn, text="Update", command=update_file)
btn_update.grid(row=2, column=0, sticky="ew", padx=5, pady=5)

window.mainloop()
```
